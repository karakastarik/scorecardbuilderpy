# Feature Engineering

```python

#Extracting limit bal ratio from PAY_AMT variable for standartization
for i in df.columns[17:23]:
    df["LIMIT_BAL_"+i]=df[i]/df.LIMIT_BAL
df["LIMIT_BAL_PAY_AMT_TOT"]=df.iloc[:,24]+df.iloc[:,25]+df.iloc[:,26]+df.iloc[:,27]+df.iloc[:,28]+df.iloc[:,29]#+df.iloc[:,30]
#one hot encoding for the below columns
df=pd.get_dummies(df, columns = ['SEX', 'EDUCATION','MARRIAGE','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6',])
```

```python
train, test = sc.split_df(df, 'default').values()
```

```python
y_train = train.loc[:,'default']
X_train = train.loc[:,train.columns != 'default']
y_test = test.loc[:,'default']
X_test = test.loc[:,train.columns != 'default']
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import neighbors
from sklearn.svm import SVR
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from warnings import filterwarnings
filterwarnings('ignore')
from sklearn.metrics import accuracy_score
import time
```

```python
modeller=[LGBMClassifier,XGBClassifier,GradientBoostingClassifier,RandomForestClassifier,DecisionTreeClassifier,KNeighborsClassifier,LogisticRegression]
def compML(algoritma):
    start = time.time()
    model=algoritma().fit(X_train,y_train)
    y_pred=model.predict(X_test)
    accuracy=accuracy_score(y_test, y_pred)
    model_ismi=algoritma.__name__
    end = time.time()
    print(model_ismi,"Accuracy:",accuracy)
for i in modeller:
    compML(i)
```

```python
# logistic regression ------
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(n_jobs=-1,verbose=5)
lr.fit(X_train, y_train)
```

```python
predicted_test=pd.Series(lr.predict(X_test),name="predicted_test").reset_index(drop=True)
predicted_train=pd.Series(lr.predict(X_train),name="predicted_train").reset_index(drop=True)

from sklearn.metrics import accuracy_score
print("Train Data Accuracy is: ",accuracy_score(y_train, predicted_train))
print("Test Data Accuracy is: ",accuracy_score(y_test, predicted_test)) 
```
